```python
import pandas as pd
from jobspy import scrape_jobs

# 1. Define your search
jobs = scrape_jobs(
    site_name=["linkedin"],           # Can also add ["indeed", "glassdoor"]
    search_term="Data Engineer",
    location="Remote",
    results_wanted=1000,                # Number of jobs to fetch
    hours_old=72,                     # Only fetch jobs posted in last 3 days
    linkedin_fetch_description=True,
    
)

# 2. Check if jobs were found
print(f"Found {len(jobs)} jobs")

jobs
```

```python
import csv

jobs.to_csv("jobs.csv", quoting=csv.QUOTE_NONNUMERIC, escapechar="\\", index=False) # to_excel
```

```python
import pandas as pd
import re

us_states = (
    r"\b(?:AL|AK|AZ|AR|CA|CO|CT|DE|FL|GA|HI|ID|IL|IN|IA|KS|KY|LA|ME|MD|MA|MI|MN|MS|MO|MT|"
    r"NE|NV|NH|NJ|NM|NY|NC|ND|OH|OK|OR|PA|RI|SC|SD|TN|TX|UT|VT|VA|WA|WV|WI|WY)\b"
)

mask_not_staffing = ~jobs["company_industry"].str.contains(
    "Staffing|Recruiting", case=False, na=False
)

mask_uk = jobs["location"].str.contains(
    r"\bUK\b|United Kingdom|England|Scotland|Wales", case=False, na=False
)

mask_us = (
    jobs["location"].str.contains("United States|USA|US", case=False, na=False)
    | jobs["location"].str.contains(us_states, regex=True, na=False)
)

filtered_df = jobs[mask_not_staffing & (mask_uk | mask_us)]
filtered_df[["id", "company_industry", "job_url", "title", "company", "location", "description", "job_function"]].head()
```
